```python
df = df.drop_duplicates()
df.shape
```

```python
n = len(df)
def column_repeat_values(dfc):
    for column in dfc:
        n_for_column = dfc[column].value_counts()
        max_value  = n_for_column.iloc[0]
        min_value = n_for_column.iloc[-1]
        print(f'{dfc[column].name} | {round(max_value / (1.0*n), 3) * 100} - {round(min_value / (1.0*n), 3) * 100} | {dfc[column].dtype}')

column_repeat_values(df)
```

```python
def have_null_value(dfc):
    n = float(len(df))
    for column in dfc:
        n_null = float(len(dfc[dfc[column].isnull()]))
        print(f'{dfc[column].name} = {(n_null/36791.0)}')
        
have_null_value(df)
```

```python
def get_z_score(dfc):
    for column in dfc:
        if dfc[column].dtypes != np.object:
            n_outliers = len(dfc[np.abs(stats.zscore(dfc[column])) > 3])
            print(f'{column} | {n_outliers} | {dfc[column].dtypes}')
            
get_z_score(df)
```

```python
def unique_value(dfc):
    for column in dfc:
        print(f'{column} | {dfc[column].unique().shape[0]} | {dfc[column].dtypes}')
        
unique_value(df)
```

```python
df.clase.unique()
```

```python
litle = ['Subcompact Cars', 'Compact Cars', 'Two Seaters', 'Minicompact Cars']
medium = ['Midsize Cars']
big = ['Large Cars']

df.loc[df.clase.isin(litle), 'clase_tipo'] = 'coche_pequeno'
df.loc[df.clase.isin(medium), 'clase_tipo'] = 'medium'
df.loc[df.clase.isin(big), 'clase_tipo'] = 'big'
```

```python
df.clase_tipo = df.clase_tipo.astype('category')
```

```python
df.clase_tipo.dtype
```

```python
df.traccion.unique()
```

```python
cuatro = ['4-Wheel or All-Wheel Drive', 'All-Wheel Drive', '4-Wheel Drive',  'Part-time 4-Wheel Drive']
df['traccion_tipo'] = 'dos'
df.loc[df.traccion.isin(cuatro), 'traccion_tipo'] = 'cuatro'
```

```python
df.traccion_tipo = df.traccion_tipo.astype('category')
```

```python
df['transmision_tipo'] = 'automatic'
df.loc[(df.transmision.notnull()) & (df.transmision.str.startswith('M')), 'transmision_tipo' ] = 'manual'
```

```python
df.combustible.unique()
```

```python
df['combustible_tipo'] = 'Othes'

df.loc[df['combustible'] == 'Regular', 'combustible_tipo'] = 'regular'
df.loc[df['combustible'] == 'Premium', 'combustible_tipo'] = 'premium'
df.loc[df['combustible'].str.contains('Electricity'), 'combustible_tipo'] = 'hybrid'
```

```python
df['combustible_tipo'] = df['combustible_tipo'].astype('category')
```
